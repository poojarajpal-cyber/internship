```python
# lets train_test and split
# for that importing necessary libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,r2_score,classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
# lets divide the dataset into label and features


x=data.drop(columns=['Country/Territory'])
y=data['Country/Territory']
```

```python
# training and testing the dataset
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=355)
```

```python
# initialising and training the model
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
pred_train=knn.predict(x_train)
pred_test=knn.predict(x_test)
```

```python
#checking the training and testing score
```

```python
print("training score is:",r2_score(y_train,pred_train))
print("testing score is:",r2_score(y_test,pred_test))
```

```python
# cross validation score to check if the model is overfitting
# for that importing necessary library
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(knn,x,y,cv=5)
```

```python
cross_val_score(KNeighborsClassifier(),x,y,cv=5).mean()
```

```python
# lets use theGridSearchCV to improve the accuracy
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'algorithm':['kd_tree','brute'],
           'leaf_size':[3,5,6,7,8],
           'n_neighbors':[3,5,7,9,11,13]}
grd_srch=GridSearchCV(estimator=knn,param_grid=param_grid)
```

```python
grd_srch.fit(x_train,y_train)
grd_srch.best_params_
```

```python
knn=KNeighborsClassifier(algorithm='kd_tree',leaf_size=3,n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
pred_train=knn.predict(x_train)
pred_test=knn.predict(x_test)
```

```python
print("training score is:",r2_score(y_train,pred_train))
print("testing score is:",r2_score(y_test,pred_test))
```
